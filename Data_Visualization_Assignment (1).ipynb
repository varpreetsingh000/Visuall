Ques-5 Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


np.random.seed(42)


num_samples = 100
x1 = np.random.rand(num_samples)
x2 = x1 * 0.8 + np.random.rand(num_samples) * 0.2
x3 = x1 * -0.5 + np.random.rand(num_samples) * 0.5
x4 = np.random.rand(num_samples)


df = pd.DataFrame({
    'Feature 1': x1,
    'Feature 2': x2,
    'Feature 3': x3,
    'Feature 4': x4
})


correlation_matrix = df.corr()


plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)


plt.title("Correlation Matrix Heatmap")


plt.show()
```

```python
import plotly.express as px


np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df_sales = pd.DataFrame(data)


heatmap_data = df_sales.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='sum')


fig = px.imshow(heatmap_data, aspect="auto", color_continuous_scale='Viridis',
                title='Heatmap of Sales Variation Across Months and Days')


fig.show()
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np


x = np.linspace(0, 10, 100)
y = np.sin(x)


output_notebook()

plot = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')
plot.line(x, y, legend_label="sin(x)", line_width=2)


show(plot)
```

Ques-2 Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

Ques-3 Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']

counts = [20, 25, 30, 35]

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

fruits = ['Apple', 'Orange', 'Banana', 'Pear']
counts = [20, 25, 30, 35]

# Create a ColumnDataSource to hold the data
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Enable notebook output for Bokeh plots
output_notebook()

# Create a bar chart
plot = figure(x_range=fruits, title="Fruit Counts",
             x_axis_label='Fruits', y_axis_label='Counts',
             height=400, width=600)
plot.vbar(x='fruits', top='counts', width=0.9, source=source,
         fill_color="navy", line_color="white")

# Add labels to the bars
for fruit, count in zip(fruits, counts):
    plot.text(x=fruit, y=count + 1, text=[str(count)],
              text_align="center")

# Show the plot
show(plot)
```

Ques-4 Create a Bokeh histogram to visualize the distribution of the given data.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np


data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)


output_notebook()

plot = figure(title="Histogram of Random Data", x_axis_label='Value', y_axis_label='Frequency', height=400, width=600)
plot.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", line_color="white", alpha=0.7)


show(plot)
```

Ques-5  Create a Bokeh heatmap using the provided dataset.

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar, ColumnDataSource
import numpy as np


data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)


data_flat = data_heatmap.flatten()
x_flat = xx.flatten()
y_flat = yy.flatten()


source = ColumnDataSource(data=dict(x=x_flat, y=y_flat, value=data_flat))


output_notebook()


plot = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y', width=400, height=400)


mapper = linear_cmap(field_name='value', palette="Viridis256", low=min(data_flat), high=max(data_flat))


plot.rect('x', 'y', width=0.1, height=0.1, source=source,
          fill_color=mapper, line_color="white")


color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0, 0))
plot.add_layout(color_bar, 'right')


show(plot)
```
